```python
# -----------------------------------------
# Sistema de Seguimiento de Progreso de Ajedrecistas
# -----------------------------------------

progreso_ajedrecistas = {}

# ---------------- Funciones ----------------

def registrar_ajedrecista():
    nombre = input("Ingrese el nombre del ajedrecista: ").strip()
    if nombre in progreso_ajedrecistas:
        print("‚ö†Ô∏è Ese ajedrecista ya est√° registrado.")
        return

    tactica = int(input("Puntaje de T√°ctica (0-100): "))
    estrategia = int(input("Puntaje de Estrategia (0-100): "))
    rapidez_visual = int(input("Puntaje de Rapidez Visual (0-100): "))

    progreso_ajedrecistas[nombre] = {
        "T√°ctica": tactica,
        "Estrategia": estrategia,
        "Rapidez Visual": rapidez_visual
    }
    print(f"‚úÖ Ajedrecista '{nombre}' registrado con √©xito.")

def mostrar_progreso():
    if not progreso_ajedrecistas:
        print("No hay ajedrecistas registrados.")
        return

    print("\nüìä Progreso de Ajedrecistas:")
    for nombre, datos in progreso_ajedrecistas.items():
        print(f"- {nombre}: T√°ctica={datos['T√°ctica']}, Estrategia={datos['Estrategia']}, Rapidez Visual={datos['Rapidez Visual']}")

def actualizar_puntaje():
    nombre = input("Ingrese el nombre del ajedrecista a actualizar: ").strip()
    if nombre not in progreso_ajedrecistas:
        print("‚ö†Ô∏è Ese ajedrecista no est√° registrado.")
        return

    tactica = int(input("Nuevo puntaje de T√°ctica (0-100): "))
    estrategia = int(input("Nuevo puntaje de Estrategia (0-100): "))
    rapidez_visual = int(input("Nuevo puntaje de Rapidez Visual (0-100): "))

    progreso_ajedrecistas[nombre] = {
        "T√°ctica": tactica,
        "Estrategia": estrategia,
        "Rapidez Visual": rapidez_visual
    }
    print(f"‚úÖ Puntajes de '{nombre}' actualizados.")

def calcular_promedios():
    if not progreso_ajedrecistas:
        print("No hay datos para calcular promedios.")
        return

    total_tactica = total_estrategia = total_rapidez = 0
    cantidad = len(progreso_ajedrecistas)

    for datos in progreso_ajedrecistas.values():
        total_tactica += datos["T√°ctica"]
        total_estrategia += datos["Estrategia"]
        total_rapidez += datos["Rapidez Visual"]

    print("\nüìà Promedios Generales:")
    print(f"T√°ctica: {total_tactica / cantidad:.2f}")
    print(f"Estrategia: {total_estrategia / cantidad:.2f}")
    print(f"Rapidez Visual: {total_rapidez / cantidad:.2f}")

def borrar_ajedrecista():
    nombre = input("Ingrese el nombre del ajedrecista a borrar: ").strip()
    if nombre in progreso_ajedrecistas:
        del progreso_ajedrecistas[nombre]
        print(f"üóëÔ∏è Ajedrecista '{nombre}' eliminado del registro.")
    else:
        print("‚ö†Ô∏è Ese ajedrecista no est√° registrado.")

# ---------------- Men√∫ Principal ----------------
def menu():
    while True:
        print("\n------ Sistema de Seguimiento de Progreso de Ajedrecistas ------")
        print("1. Registrar ajedrecista")
        print("2. Mostrar progreso")
        print("3. Actualizar puntajes")
        print("4. Calcular promedios")
        print("5. Salir")
        print("6. Borrar ajedrecista")  # Nueva opci√≥n

        opcion = input("Seleccione una opci√≥n: ")

        if opcion == "1":
            registrar_ajedrecista()
        elif opcion == "2":
            mostrar_progreso()
        elif opcion == "3":
            actualizar_puntaje()
        elif opcion == "4":
            calcular_promedios()
        elif opcion == "5":
            print("üëã Saliendo del sistema...")
            break
        elif opcion == "6":
            borrar_ajedrecista()
        else:
            print("‚ö†Ô∏è Opci√≥n no v√°lida, intente de nuevo.")

# Ejecutar el programa
menu()
```
